# Replace

```python
import pandas as pd
r=pd.read_csv("F:\\Data Science(1182)\\DS Youtube\\Pandas\\student.csv")
r
```

```python
r.replace(to_replace="female",value="trans")
```

```python
r.replace([1,2,3,4,5,6,7],22)
```

```python
r.replace("[A-Za-z]","python",regex=True)
```

```python
r.replace("[A-Z]","python",regex=True)
```

```python
r.replace({"name":'[A-Z]'},22,regex=True)
```

```python
r.replace("male",method="bfill")
```

```python
r.replace(88,method="ffill",limit=1)
```

```python
r.replace(88,method="ffill",limit=1,inplace=True)
r
```

# Interpolate

```python
import pandas as pd
r=pd.read_csv("F:\\Data Science(1182)\\DS Youtube\\Pandas\\student2.csv")
r
```

```python
r.interpolate() # it only work on numeric value
```

```python
r.interpolate(method="linear")
```

```python
r.interpolate(method="linear",axis=0) # for axis 1 error showing datatype must be same
```

```python
r.interpolate(limit=2)
```

```python
r.interpolate(limit_direction="both",limit=2) # you can you backward, forward at the place of both
```

```python
r.interpolate(limit_area="outside") # inside also u can use
```

```python
r.interpolate(limit_direction="both",limit=2,inplace=True)
r
```
