# Merge

```python
import pandas as pd
a=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
b=pd.DataFrame({"A":[1,2,3,5],"C":[21,22,23,24]})
pd.merge(a,b,on="A") # depending on common data it will give output
```

```python
pd.merge(a,b,how="right") #left,right,inner,outer also there 
```

```python
pd.merge(a,b,how="right",indicator=True) 
```

```python
a=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
b=pd.DataFrame({"A":[1,2,3,5],"B":[21,22,23,24]})
pd.merge(a,b)
```

```python
pd.merge(a,b,left_index=True,right_index=True)
```

```python
pd.merge(a,b,left_index=True,right_index=True,suffixes=("name","id"))
```

# Concat

```python
s1=pd.Series([1,2,3,4])
s2=pd.Series([11,21,13,14])

pd.concat([s1,s2])
```

```python
d1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
d2 =pd.DataFrame({"A":[1,2,3,5],"B":[21,22,23,24]})
pd.concat([d1,d2])
```

```python
pd.concat([d1,d2],axis=1)
```

```python
d1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
d2 =pd.DataFrame({"A":[1,2],"C":[23,24]})
pd.concat([d1,d2],axis=1,join="inner")
```

```python
d1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
d2 =pd.DataFrame({"A":[1,2,3,5],"B":[21,22,23,24]})
pd.concat([d1,d2],axis=1,keys=["d1","d2"])
```

```python
d1=pd.DataFrame({"V":[1,2,3,4]})
d2 =pd.DataFrame({"A":[1,2,3,5],"B":[21,22,23,24]})
pd.concat([d1,d2])
```
