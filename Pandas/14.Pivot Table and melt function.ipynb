```python
import pandas as pd
m=pd.DataFrame({"days":[1,2,3,4,5,6],"eng":[10,12,14,15,17,18],"maths":[17,18,16,14,23,45]})
m
```

```python
pd.melt(m)
```

```python
pd.melt(m,id_vars=["eng"])
```

```python
pd.melt(m,id_vars=["eng"],var_name="python",value_name="JBK")
```

# Pivot

```python
import pandas as pd
p=pd.DataFrame({"days":[1,2,3,4,5,6],"st_name":["a","b","c","a","b","c"],"eng":[10,12,14,15,17,18],"maths":[17,18,16,14,23,45]})
p
```

```python
p.pivot(index="days",columns="st_name")
```

```python
p.pivot(index="days",columns="st_name",values="eng")
```

```python
p=pd.DataFrame({"days":[1,1,1,1,2,2],"st_name":["a","b","a","a","b","b"],"eng":[10,12,14,15,17,18],"maths":[17,18,16,14,23,45]})
p.pivot_table(index="st_name",columns="days",aggfunc="mean") # u can use sum,count
```

```python
p.pivot_table(index="st_name",columns="days",aggfunc="mean",margins="True") #margins show average value
```
