# Scatter

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[2,1,4,2,6,3,8]
plt.scatter(day,no,color="r")
plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No",fontsize=15)
plt.show()
```

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
no1=[3,2,1,6,5,4,7]
color=["r","y","g","b","r","g","r"]
sizes=[400,330,450,690,440,560,100]
plt.scatter(day,no,c=color,s=sizes,alpha=0.5,marker="*",edgecolor="y",linewidth=3)
plt.scatter(day,no1,color="r",s=sizes,alpha=0.5,marker="*",edgecolor="y",linewidth=3)

plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No",fontsize=15)
plt.show()
```

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[2,1,4,2,6,3,8]
color=[10,49,39,29,45,55,56]
sizes=[400,200,400,300,200,100,600]
plt.scatter(day,no,c=color,s=sizes,cmap="viridis")
plt.title("Scatter Plot",fontsize=15)
t=plt.colorbar()
t.set_label("ColorBar",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No",fontsize=15)
plt.show()
```
