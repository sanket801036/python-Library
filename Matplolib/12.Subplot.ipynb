```python
import matplotlib.pyplot as plt
import numpy as np

# Create some data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(x)

# Create a figure with 2x2 subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Plot on the first subplot (top-left)
axs[0, 0].plot(x, y1, 'r')  # 'r' stands for red color
axs[0, 0].set_title('Sine Wave')
axs[0, 0].set_xlabel('x') # [0, 0]: These are the indices used to access a specific subplot. In this case,
# it refers to the subplot located in the first row and the first column.
axs[0, 0].set_ylabel('sin(x)')

# Plot on the second subplot (top-right)
axs[0, 1].plot(x, y2, 'g')  # 'g' stands for green color
axs[0, 1].set_title('Cosine Wave')
axs[0, 1].set_xlabel('x')
axs[0, 1].set_ylabel('cos(x)')

# Plot on the third subplot (bottom-left)
axs[1, 0].plot(x, y3, 'b')  # 'b' stands for blue color
axs[1, 0].set_title('Tangent Wave')
axs[1, 0].set_xlabel('x')
axs[1, 0].set_ylabel('tan(x)')
axs[1, 0].set_ylim(-10, 10)  # Limit y-axis to avoid extreme values

# Plot on the fourth subplot (bottom-right)
axs[1, 1].plot(x, y4, 'k')  # 'k' stands for black color
axs[1, 1].set_title('Exponential Function')
axs[1, 1].set_xlabel('x')
axs[1, 1].set_ylabel('exp(x)')
axs[1, 1].set_ylim(0, 22000)  # Limit y-axis to avoid extreme values

# Adjust layout to prevent overlapping
fig.tight_layout()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create some data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create a figure with 2 subplots (1 row, 2 columns)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

# Plot on the first subplot
ax1.plot(x, y1, 'r')  # 'r' stands for red color
ax1.set_title('Sine Wave')
ax1.set_xlabel('x')
ax1.set_ylabel('sin(x)')

# Plot on the second subplot
ax2.plot(x, y2, 'g')  # 'g' stands for green color
ax2.set_title('Cosine Wave')
ax2.set_xlabel('x')
ax2.set_ylabel('cos(x)')

# Adjust layout to prevent overlapping
fig.tight_layout()

# Show the plot
plt.show()
```
