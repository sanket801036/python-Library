```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=["c","c++","Java","python"]
ex=[0.4,0.0,0.0,0.0]
c=["r","b","g","y"]
plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.2f%%",shadow=True,radius=1,labeldistance=1.2,startangle=120,textprops={"fontsize":15},counterclock=False,
        wedgeprops={"linewidth":5,"edgecolor":"m"},rotatelabels=False)
plt.title("JBK Tech")
plt.legend(loc=2)
plt.show()
```
