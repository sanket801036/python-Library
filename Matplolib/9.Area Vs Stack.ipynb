```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
area1 = [2, 3, 2, 5, 4]
area2 = [2, 3, 4, 5, 6]
area3 = [1, 3, 2, 4, 2]
l = ["area1", "area2", "area3"]  # Changed labels to match area names

# Corrected the function call
# plt.stackplot(x, area1, area2, area3, labels=l, colors=["r", "g", "m"],baseline="wiggle")
plt.stackplot(x, area1, area2, area3, labels=l, colors=["r", "g", "m"])
plt.legend(loc=2)
plt.title("Stacked Area Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```
