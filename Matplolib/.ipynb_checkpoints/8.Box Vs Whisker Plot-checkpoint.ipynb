```python
import matplotlib.pyplot as plt

x=[10,20,30,40,50,60,70,120]
# plt.boxplot(x,notch=False)
# plt.boxplot(x,vert=False)
# plt.boxplot(x,widths=0.5)
# plt.boxplot(x,label=["python"],patch_artist=True)
# plt.boxplot(x,label=["python"],showmeans=True,whis=2)
plt.boxplot(x,label=["python"],showmeans=True,sym="g",boxprops=dict(color="r"),capprops=dict(color="g"),whiskerprops=dict(color="r")
           ,flierprops=dict(markeredgecolor="y")) # sym=" ":- it will make gone that dot
plt.show()
```

```python
x=[10,20,30,40,50,60,70]
x1=[10,20,40,50,30,60,90]
y=[x,x1]
plt.boxplot(y,labels=["python","c++"]
            ,showmeans=True,sym="g",boxprops=dict(color="r"),capprops=dict(color="g"),whiskerprops=dict(color="r")
           ,flierprops=dict(markeredgecolor="y")) # sym=" ":- it will make gone that dot
plt.show()
```
