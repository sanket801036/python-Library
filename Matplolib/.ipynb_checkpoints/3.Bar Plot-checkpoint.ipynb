```python
import matplotlib.pyplot as plt
x=["python","C","c++","java"]
y=[85,80,60,82]
plt.xlabel("language",fontsize=20)
plt.ylabel("No. ",fontsize=20)
plt.title("wscube",fontsize=20)
c=["y","b","m","g"]
plt.bar(x,y,width=0.4,color=c,align="center",edgecolor="r",linewidth=3,linestyle=":",alpha=0.5) # align="edge"
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=["python","C","c++","java"]
y=[85,80,60,82]
z=[20,30,40,50]
width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel("language",fontsize=20)
plt.ylabel("No. ",fontsize=15)
plt.title("JBK",fontsize=15)

plt.bar(p,y,width,color="r",label="popularity1") 
plt.bar(p1,z,width,color="y",label="popularity1") 
plt.xticks(p+width/2,x,rotation=20)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=["python","C","c++","java"]
y=[85,80,60,82]
z=[20,30,40,50]
width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel("language",fontsize=20)
plt.ylabel("No. ",fontsize=15)
plt.title("JBK",fontsize=15)

plt.barh(p,y,width,color="r",label="popularity1") 
plt.barh(p1,z,width,color="y",label="popularity1") 
plt.xticks(p+width/2,x,rotation=20)
plt.legend()
plt.show()
```
