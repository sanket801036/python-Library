```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[2,4,3,6,4]
plt.plot(x,y,color="r")
plt.savefig("line1",dpi=2000, facecolor="g")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[2,4,3,6,4]
plt.plot(x,y,color="r")
plt.savefig("line45.pdf",dpi=2000, facecolor="g",transparent=False, bbox_inches="tight")
plt.show()
```
