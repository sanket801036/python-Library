```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

var = pd.read_csv('tips.csv') 
var
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=var,hue="sex",color="r",order=["Fri","Sun","Thur","Sat"],showmeans=True,meanprops={"marker":"+","markeredgecolor":"b"}
           ,linewidth=3,palette="plasma",orient="v")
plt.show()
```
