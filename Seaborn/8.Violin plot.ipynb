## Violin Plot in Seaborn

**Explanation:**

* A violin plot is similar to a box and whisker plot and shows the distribution of quantitative data across various levels of a categorical variable. This helps compare the distributions. 

**Implementation:**

*  **Import Seaborn:**  `import seaborn`
* **Set style to 'whitegrid':**  `seaborn.set(style = 'whitegrid')`
* **Load the dataset:** `f = seaborn.load_dataset("fmri")`
* **Create the violin plot:** `seaborn.violinplot(x ="",y ="",data = f)` 


```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

var = pd.read_csv('tips.csv') 
var
```

```python
sns.violinplot(x="day",y="total_bill",data=var,hue="time")
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=var,linewidth=2,palette='Dark2',saturation=0.4)

plt.show()
```

```python
sns.violinplot(x="day", y="total_bill", data=var, order=["Thur", "Fri", "Sat", "Sun"],color="y"
               ,inner="stick")#{"box","quartile","stick",None})

# Show the plot
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=var,hue="sex",split=True,scale="width")
plt.show()
```

```python
sns.violinplot(x=var["total_bill"])
plt.show()
```
