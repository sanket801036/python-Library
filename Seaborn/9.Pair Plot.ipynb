```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

var = pd.read_csv('tips.csv') 
var
```

```python
sns.pairplot(var)
plt.show()
```

```python
sns.pairplot(var,vars=["total_bill","tip"],hue="sex",hue_order=["Female","Male"])
plt.show()
```

```python
sns.pairplot(var,hue="sex",hue_order=["Female","Male"],x_vars=["total_bill","tip"],markers=["*","<"])
plt.show()
```

```python
sns.pairplot(var,hue="sex",hue_order=["Female","Male"],kind="reg") # 'scatter','kde','hist','reg'
plt.show()
```
