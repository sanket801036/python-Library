```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

var = pd.read_csv('penguins.csv') 
var
```

```python
markers={"MALE":"*","FEMALE":"o"}
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=var,hue="sex",style="sex",size="sex",sizes=(100,60),palette="gist_ncar",alpha=1,markers=markers)
plt.show()
```
